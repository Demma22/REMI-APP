```python
# === STEP 1: 
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import joblib
```

```python
#Load data
df = pd.read_csv("ugandan_students.csv")

#Train vectorizer on student names
vectorizer = TfidfVectorizer(analyzer='char_wb', ngram_range=(2,4))
X = vectorizer.fit_transform(df['Name'])
```

```python
#Define a function to find best match
def find_best_match(query):
    query_vec = vectorizer.transform([query])
    sims = cosine_similarity(query_vec, X).flatten()
    idx = sims.argmax()
    result = df.iloc[idx]
    return result.to_dict(), sims[idx]
```

```python
#Test
query = "Samara"
match, score = find_best_match(query)
print("Top match:", match)
print("Confidence:", score)
```

```python
import joblib

joblib.dump(vectorizer, "name_vectorizer.pkl")
df.to_csv("students_db.csv", index=False)
```
